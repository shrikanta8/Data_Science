```python
# Merge operations
```

```python
data1 = pd.DataFrame({
    'key1': [3,4,5,6],
    'key2': [36,7,8,9],
    'key3': [9,0,2,3]
})
```

```python
data2 = pd.DataFrame({
    'key5': [7,8,9,0],
    'key4': [2,3,4,6],
    'key3': [8,1,2,3]
})
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1, data2)   # by default it is inner merge
```

```python
# it take the common values(2,3) of common columns(key3).
# with that it prints all the columns values which are there in the particular row of key3((2,3)values)
```

```python
pd.merge(data1, data2, how='left')
```

```python
pd.merge(data1, data2, how='right')
```

```python
pd.merge(data1, data2, how='outer')    #it will bring whole
```

```python
pd.merge(data1, data2, how='outer',on='key3')    # 'on' tells on which column we have to merge
```

```python
pd.merge(data1, data2, how='cross')  #cartesian product
```

```python
# join operations on row index
```

```python
data1 = pd.DataFrame({
        'key1': [3,4,5,6],
        'key2': [36,7,8,9],
        'key3': [9,0,2,3]
    },
    index = ['a','b','c','d']
)
```

```python
data1
```

```python
data2 = pd.DataFrame({
        'key5': [7,8,9,0],
        'key4': [2,3,4,6],
        'key13': [8,1,2,3]
    },
    index = ['a','b','g','h']
)
```

```python
data2
```

```python
data1.join(data2)  #by default it joins in left
```

```python
data1.join(data2,how='right') 
```

```python
data1.join(data2,how='inner') 
```

```python
data1.join(data2,how='outer') 
```

```python
data1.join(data2,how='cross') 
```

```python
# add a new column having INR values of fare
```

```python
df['fare_INR'] = df['Fare'].apply(lambda x: x*x*80)
```

```python
df.head()
```

```python
def euro_convert(x):
    return x+8

df['Age_update'] = df['Age'].apply(euro_convert)
```

```python
df.head()
```

```python
df['name_len'] = df['Name'].apply(len)
```

```python
df.head()
```

```python
def chk(x):
    if x<10:
        return "cheap"
    elif x>=10 and x<20:
        return "mid"
    else:
        return "high"

df['Fare_chk'] = df['Fare'].apply(chk)
```
