```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
# to get how x and y are moving
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("this is x data")
plt.ylabel("this is y data")
plt.title("this is titlee")
plt.show()
```

```python
plt.figure(figsize=(6,2))
plt.plot(x,y,'--r')      # dashed red line
plt.xlabel("this is x data")
plt.ylabel("this is y data")
plt.title("this is titlee")
plt.grid()     # background grid
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
```

```python
x
```

```python
y
```

```python
colours = np.random.rand(50)
sz = np.random.rand(50)*1000
plt.scatter(x,y,c = colours,s=sz,alpha=.4)   # alpha is colour density
plt.xlabel("x data")
plt.ylabel("y data")
plt.title("random data")
```

```python
sz = np.random.rand(50)*1000
plt.scatter(x,y,c = 'b',s=sz,alpha=.4, marker='v')   
# plt.scatter(x,y,c = '#33FF6E',s=sz,alpha=.4, marker='x')  
plt.xlabel("x data")
plt.ylabel("y data")
plt.title("random data")
```
